```python
import numpy as np
vocabulary={}
labels=[]

with open('Pickle/SMSSpamCollection1', encoding="utf-8") as file_handle:
    for line in file_handle: #파일을 한 줄씩 읽는다.
        splits=line.split() # 한 줄을 단어별로 쪼개서 리스트로 만든다.
        label=splits[0] # 맨 앞의 단어는 레이블로 따로 둔다.
        if label == 'spam': # 맨 앞 단어가 spam이면 1, 아니면 0을 추가한다.
            labels.append(1)
        else:
            labels.append(0)
        
        text=splits[1:]
        # 전체 내용을 단어 단위로 살펴본다.
        # 사전에 해당 단어가 없으면 추가해서 고유번호를 붙인다.
        # 그리고 그 매핑을 vocabulary에 저장한다.
        for word in text:
            lower=word.lower() # 영문자를 소문자로 통일한다.
            if not lower in vocabulary:
                vocabulary[lower]=len(vocabulary)
```

```python
print(len(labels)) #문서가 총 5574개 있다.
```

```python
print(vocabulary) #단어집에 존재하는 단어수는 13627개이다.
```

```python
#단어빈도피처 만들기
features=[]
with open('Pickle/SMSSpamCollection1',encoding="utf-8") as file_handle:
    for line in file_handle: #파일을 한 줄씩 읽는다.
        feature=np.zeros(len(vocabulary)) # 0으로 채워진 numpy 벡터를 만든다.
        splits=line.split() 
        text=splits[1:]
        for word in text:
            lower=word.lower() # 영문자를 소문자로 통일한다.
            feature[vocabulary[lower]]+=1
            print(feature)
            
        feature=feature/sum(feature)
        features.append(feature)
```

```python
print(features) #단어집에 존재하는 단어 수보다 문서에 나오는 단어 수가 월등하게 작기 때문에 대부분 0으로 차 있게 된다.
```

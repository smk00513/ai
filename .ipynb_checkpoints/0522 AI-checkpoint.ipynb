```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("Image/talmo.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imgray, cmap = "gray")
plt.show()
```

```python
templ = cv2.imread("Image/talmoflower.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(templ, cmap = "gray")
plt.show()
```

```python
print(templ.shape)
h,w = templ.shape[:2]
```

```python
res = cv2.matchTemplate(imgray, templ, cv2.TM_CCOEFF_NORMED)
plt.imshow(res, cmap = "gray")
plt.show()
```

```python
loc = np.where(res>= 0.4)
print(loc)
```

```python
for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0]+w, pt[1]+h), (0,0,255), 1)
```

```python
cv2.imshow("find.png", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
